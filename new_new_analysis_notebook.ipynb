```python
def get_npg_ratio(profs, ratio=0.61):
    f10 = profs['Re(freq)'][1]
    return profs.iloc[np.argmin([np.abs(f10/pl - ratio) for pl in profs['Re(freq)']])]['n_pg']

def get_idx_ratio(profs, ratio=0.61, ell=None):
    f10 = profs['Re(freq)'][1]
    if ell is None:
        return np.argmin([np.abs(f10 / pl - ratio) for pl in profs['Re(freq)']])
    
    else:
        # ps = [p for p in profs[profs['l']==8]]
        ps = profs[profs['l']==ell]
        # print([pl*f['Re(freq)'][1] for pl in ps['Re(freq)']])

        return np.argmin([np.abs(f10/ pl - ratio) for pl in ps['Re(freq)']])

# period(x)/period(first overtone) ~ 0.61  = 1/frequency(x) / 1/frequency(overtone) = frequency(overtone)/frequency(x)


# pl/p10 = pl * f10
```

```python
def plot_period_spacing_notmodel(freqs, gyre, xlim=None, ylim=None):
    freqs.n_g = [int(x) for x in freqs.n_g]
        # try:
    dipole_g = freqs[np.logical_and(freqs.l == 1, freqs.n_g > 0)] 
        # except:
            # sometimes the numbers are strings L
            # dipole_g = freqs[np.logical_and(freqs.l == 1, int(freqs.n_g)]

    dipole_g['P']  = 1/(dipole_g['Re(freq)'] * 10**-6) # seconds


    for mode in dipole_g.iterrows():
            # n_g  = mode[1]['n_g']
            # n_g2 = dipole_g[dipole_g['n_g'] == n_g+1]
            # if not n_g2.empty:
            #     dP = (n_g2['P'].values[0] - mode[1]['P']) # period spacing in seconds 
            #     dipole_g.loc[mode[0], 'dP'] = dP
        n_pg = mode[1]['n_pg']
        n_pg2 = dipole_g[dipole_g['n_pg'] == n_pg-1]

        if not n_pg2.empty:
            dP = (n_pg2['P'].values[0] - mode[1]['P']) # seconds
            dipole_g.loc[mode[0], 'dP'] = dP

    def get_Pg():
        N  = gyre.N
        x  = gyre.x
        delta_Pg = 2*np.pi**2/np.sqrt(2)/ sp.integrate.trapezoid(N[N>0]/x[N>0], x[N>0])

        return delta_Pg
        
    plt.plot(dipole_g['P']/3600, dipole_g['dP'], 
            'b.--', ms=10, zorder=10, ls=':', mec='k', mfc='k', label='exact')
    plt.axhline(get_Pg(), ls='--', c='k', label='asymptotic')
    plt.xlim(xlim)
    plt.ylim(ylim)

    plt.xlabel('Period (hours)')
    plt.ylabel('Period Spacing (s)')
```
